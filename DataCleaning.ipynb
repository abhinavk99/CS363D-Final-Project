```python
import pandas as pd
import numpy as np
```

```python
file_name = 'H-2B_Disclosure_Data_FY2019.xlsx'
raw_data = pd.read_excel(file_name, header=0)

raw_data.head(10)
```

First, let's only keep a subset of the columns as many of the columns seem to be irrelevant in classification.

```python
RELEVANT_COLUMNS = [
    'CASE_STATUS',
    'AGENT_POC_EMP_REP_BY_AGENT',
    'SOC_CODE',
    'NAICS_CODE',
    'NBR_WORKERS_REQUESTED',
    'FULL_TIME_POSITION',
    'NATURE_OF_TEMPORARY_NEED',
    'BASIC_NUMBER_OF_HOURS',
    'BASIC_RATE_OF_PAY',
    'SUPERVISE_OTHER_EMP', # Need temporarily to fill missing values in SUPERVISE_HOW_MANY with 0
    'SUPERVISE_HOW_MANY',
    'EDUCATION_LEVEL',
    'TRAINING_REQUIRED', # Need temporarily to fill missing values in NUM_OF_MONTHS_TRAINING with 0
    'NUM_OF_MONTHS_TRAINING',
    'EMP_EXPERIENCE_REQD', # Need temporarily to fill missing values in EMP_EXP_NUM_MONTHS with 0
    'EMP_EXP_NUM_MONTHS',
    'WORKSITE_STATE',
    'SWA_NAME',
    'AGENT_ATTORNEY_CITY',
    'WORKSITE_CITY',
    'AGENT_ATTORNEY_STATE',
    'HOURLY_WORK_SCHEDULE_AM',
    'HOURLY_WORK_SCHEDULE_PM',
    'OVERTIME_RATE_FROM',
    'OVERTIME_RATE_TO',
]

df = raw_data.loc[:, RELEVANT_COLUMNS]
df.head()
```

```python
print('Number of rows before dropping rows:', df.shape[0])
```

The columns `SUPERVISE_HOW_MANY`, `NUM_OF_MONTHS_TRAINING`, and `EMP_EXP_NUM_MONTHS` contain missing values when the columns `SUPERVISE_OTHER_EMP`, `TRAINING_REQUIRED`, and `EMP_EXPERIENCE_REQD` have the value **No** respectively, so we first drop rows when there are missing values in `SUPERVISE_OTHER_EMP`, `TRAINING_REQUIRED`, and `EMP_EXPERIENCE_REQD` and fill the values in `SUPERVISE_HOW_MANY`, `NUM_OF_MONTHS_TRAINING`, and `EMP_EXP_NUM_MONTHS` with **0** when there is a **No** in `SUPERVISE_HOW_MANY`, `NUM_OF_MONTHS_TRAINING`, and `EMP_EXP_NUM_MONTHS`.

```python
df.dropna(subset=['SUPERVISE_OTHER_EMP', 'TRAINING_REQUIRED', 'EMP_EXPERIENCE_REQD'], inplace=True)
df.loc[df['SUPERVISE_OTHER_EMP'] == 'No', 'SUPERVISE_HOW_MANY'] = 0
df.loc[df['TRAINING_REQUIRED'] == 'No', 'NUM_OF_MONTHS_TRAINING'] = 0
df.loc[df['EMP_EXPERIENCE_REQD'] == 'No', 'EMP_EXP_NUM_MONTHS'] = 0
df.head()
```

Now that we've filling in missing data in `SUPERVISE_HOW_MANY`, `NUM_OF_MONTHS_TRAINING`, and `EMP_EXP_NUM_MONTHS`, we can drop the columns `SUPERVISE_OTHER_EMP`, `TRAINING_REQUIRED`, and `EMP_EXPERIENCE_REQD` and drop rows based on whether there are missing values in the row.

```python
df.drop(columns=['SUPERVISE_OTHER_EMP', 'TRAINING_REQUIRED', 'EMP_EXPERIENCE_REQD'], inplace=True)
print('Number of columns after dropping columns:', df.shape[1])
print('Number of rows after dropping columns:', df.shape[0])
df.head()
```

No rows have been dropped so far after cleaning `SUPERVISE_HOW_MANY`, `NUM_OF_MONTHS_TRAINING`, and `EMP_EXP_NUM_MONTHS`.

For the column `SWA_NAME`, we replace missing values with 0 and all other values with 1.
This is done as a feat of feature engineering, converting this sparse categorical feature into a 0 if this job is not associated with a State Workforce Agency, and converting it to 1 if the job is associated with a State Workforce Agency.

```python
df.loc[df['SWA_NAME'].isna(), 'SWA_NAME'] = 0
df.loc[df['SWA_NAME'] != 0, 'SWA_NAME'] = 1
df['SWA_NAME'] = pd.to_numeric(df['SWA_NAME'])
print(df['SWA_NAME'].value_counts())
df.head()
```

For the column `OVERTIME_RATE_FROM`, we replace missing values with 0. Then for the column `OVERTIME_RATE_TO`, we replace missing values with the value in `OVERTIME_RATE_FROM`.

```python
df.loc[df['OVERTIME_RATE_FROM'].isna(), 'OVERTIME_RATE_FROM'] = 0
df.loc[df['OVERTIME_RATE_TO'].isna(), 'OVERTIME_RATE_TO'] = df['OVERTIME_RATE_FROM']
df.head()
```

Now, let's do some standardization on the class labels, indicated by the column `CASE_STATUS`. We will consider 'Determination Issued - Certification' as 'Certified' (numerical value of 1), and 'Determination Issued - Denied' and 'Determination Issued - Rejected' as 'Denied' (numerical value of 0), dropping all other records with differing class labels.

```python
df.loc[df['CASE_STATUS'] == 'Determination Issued - Certification', 'CASE_STATUS'] = 1
df.loc[df['CASE_STATUS'] == 'Determination Issued - Denied', 'CASE_STATUS'] = 0
df.loc[df['CASE_STATUS'] == 'Determination Issued - Rejected', 'CASE_STATUS'] = 0
df = df[df['CASE_STATUS'].isin([1, 0])]
print("We retain {} rows after class label standardization.".format(df.shape[0]))
df['CASE_STATUS'].value_counts()
```

Next, we drop all rows that have missing values and all rows that are duplicates.

```python
print('AGENT_POC_EMP_REP_BY_AGENT VALUE COUNTS')
print(df['AGENT_POC_EMP_REP_BY_AGENT'].value_counts())
# So before we go, 6237 YES, 506 NO.

df_with_nan = df[df.isna().any(axis=1)]
print(len(df_with_nan))
# print(df_with_nan.head())

# We can see here that NaN removal is decimating the "no attorney" group since they don't have a city (AGENT_ATTORNEY_CITY) / state (AGENT_ATTORNEY_STATE) associated with them.
# Let's define custom values to use here to avoid the problem of NaN.
NO_ATTORNEY_REPRESENTATION = "NO ATTORNEY REPRESENTATION"

# Just to check that it's not currently in use:
assert NO_ATTORNEY_REPRESENTATION not in df['AGENT_ATTORNEY_CITY'].values
assert NO_ATTORNEY_REPRESENTATION not in df['AGENT_ATTORNEY_STATE'].values

# print(df['AGENT_ATTORNEY_CITY'].value_counts())
df.loc[df['AGENT_POC_EMP_REP_BY_AGENT'] == "No", 'AGENT_ATTORNEY_CITY'] = "NO ATTORNEY REPRESENTATION"
df.loc[df['AGENT_POC_EMP_REP_BY_AGENT'] == "No", 'AGENT_ATTORNEY_STATE'] = "NO ATTORNEY REPRESENTATION"

df_with_nan = df[df.isna().any(axis=1)]
print(len(df_with_nan))

# So we can see that we've eliminated a lot of the previously NaN rows.
# Let's see if we can do a few more.

df.loc[df['EMP_EXP_NUM_MONTHS'].isna(), 'EMP_EXP_NUM_MONTHS'] = 0
df.loc[df['SUPERVISE_HOW_MANY'].isna(), 'SUPERVISE_HOW_MANY'] = 0
df_with_nan = df[df.isna().any(axis=1)]
print(len(df_with_nan))

# Diminishing returns beyond this point, so let's stop here.
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
# TODO: Is dropping rows that are duplicates correct? Possible for multiple H2B entries to be similar, and both accepted, may not necessarily be errorneous data.
print('Number of rows after dropping rows with missing values / duplicates:', df.shape[0])
df['AGENT_POC_EMP_REP_BY_AGENT'].value_counts()
```

In columns that have Yes/No values, we replace the values with 1/0.

```python
df['AGENT_POC_EMP_REP_BY_AGENT'] = df['AGENT_POC_EMP_REP_BY_AGENT'].map({'Yes': 1, 'No': 0})
df['FULL_TIME_POSITION'] = df['FULL_TIME_POSITION'].map({'Y': 1, 'N': 0})
df.head()
```

Then, we standardize all of the string type categorical value columns to be lowercase.

```python
df['NATURE_OF_TEMPORARY_NEED'] = df['NATURE_OF_TEMPORARY_NEED'].str.lower()
df['EDUCATION_LEVEL'] = df['EDUCATION_LEVEL'].str.lower()
df['WORKSITE_STATE'] = df['WORKSITE_STATE'].str.lower()
df['AGENT_ATTORNEY_CITY'] = df['AGENT_ATTORNEY_CITY'].str.lower()
df['WORKSITE_CITY'] = df['WORKSITE_CITY'].str.lower()
df['AGENT_ATTORNEY_STATE'] = df['AGENT_ATTORNEY_STATE'].str.lower()
df.head()
```

We now standardize the column `SOC_CODE` to store integers that are 6 digits long and the column `NAICS_CODE` to store integers.

```python
df['SOC_CODE'] = df['SOC_CODE'].str.replace('-', '', regex=False)
df['SOC_CODE'] = pd.to_numeric(df['SOC_CODE'])
df['SOC_CODE'] = df['SOC_CODE'].astype('int64')
df['NAICS_CODE'] = df['NAICS_CODE'].astype('int64')
df.head()
```

```python
print('Number of rows:', df.shape[0])
print('Number of columns:', df.shape[1])
```

Now that data cleaning is completed, we output the dataframe to a new CSV.

```python
df.to_csv('H-2B_Cleaned_Data.csv', index=False)
```
