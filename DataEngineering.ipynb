```python
file_name = 'H-2B_Cleaned_data.csv'
df = pd.read_csv(file_name)
print(df.shape)
df.head(10)
```

Let's do some feature engineering on the data. 
To start with, let's create some new features from existing features that might be useful in classifying cases as certified or denied.

- [STATE_MATCH]: WORKSITE_STATE matches AGENT_ATTORNEY_STATE.
- [CITY_MATCH]: WORKSITE_CITY matches AGENT_ATTORNEY_CITY.

```python
# There has got to be a better way to do this...
df["STATE_MATCH"] = 0
df.loc[df["WORKSITE_STATE"] == df["AGENT_ATTORNEY_STATE"], "STATE_MATCH"] = 1
df.loc[df["WORKSITE_STATE"] != df["AGENT_ATTORNEY_STATE"], "STATE_MATCH"] = 0
# df.head()
# df["STATE_MATCH"].value_counts()

df["CITY_MATCH"] = 0
df.loc[df["WORKSITE_CITY"] == df["AGENT_ATTORNEY_CITY"], "CITY_MATCH"] = 1
df.loc[df["WORKSITE_CITY"] != df["AGENT_ATTORNEY_CITY"], "CITY_MATCH"] = 0
# df["CITY_MATCH"].value_counts()

# Maybe for analysis, match where statematch and citymatch are same value?
```
